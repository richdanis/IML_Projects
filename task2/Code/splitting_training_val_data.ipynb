```python
import pandas as pd
import numpy as np
```

```python
train_str = 'Data/train_features_Sven_short.csv'
labels_str = 'Data/train_labels_sorted.csv'
df_train = pd.read_csv(train_str)
df_labels = pd.read_csv(labels_str)
```

```python
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', None)
```

```python
k = np.array(df_train.pid)
pid_in_test = np.random.choice(k, size=3000, replace=False)
pid_in_test.sort()
```

```python
np_try = df_train.pid.to_list()
pid_in_test = list(pid_in_test)
pid_in_train = []
for k in np_try:
    if not k in pid_in_test:
        pid_in_train.append(k)
```

```python
np_test = np.empty((len(pid_in_test),df_train.shape[1]))
np_test_label = np.empty((len(pid_in_test),df_labels.shape[1]))

for i in range(len(pid_in_test)):
    mask_train = df_train.loc[:,'pid'] == pid_in_test[i]
    mask_label = df_labels.loc[:,'pid'] == pid_in_test[i]
    np_test[i,:] = df_train[mask_train]
    np_test_label[i,:] = df_labels[mask_label]
    
df_train_test = pd.DataFrame(np_test, columns=df_train.columns)
df_train_test = df_train_test.astype({'Age':'int32'})
df_train_test = df_train_test.astype({'pid':'int32'})
df_train_test_label = pd.DataFrame(np_test_label, columns=df_labels.columns)
df_train_test_label = df_train_test_label.astype({'pid':'int32'})
```

```python
np_train = np.empty((df_train.shape[0]-len(pid_in_test),df_train.shape[1]))
np_train_label = np.empty((df_labels.shape[0]-len(pid_in_test),df_labels.shape[1]))

for i in range(len(pid_in_train)):
    mask_train = df_train.loc[:,'pid'] == pid_in_train[i]
    mask_label = df_labels.loc[:,'pid'] == pid_in_train[i]
    np_train[i,:] = df_train[mask_train]
    np_train_label[i,:] = df_labels[mask_label]
    
df_train_train = pd.DataFrame(np_train, columns=df_train.columns)
df_train_train = df_train_train.astype({'Age':'int32'})
df_train_train = df_train_train.astype({'pid':'int32'})
df_train_train_label = pd.DataFrame(np_train_label, columns=df_labels.columns)
df_train_train_label = df_train_train_label.astype({'pid':'int32'})  
```

```python
df_train_test.to_csv('Data/train_features_test.csv', index=False, float_format='%.3f')
df_train_test_label.to_csv('Data/train_labels_test.csv', index=False, float_format='%.3f')
```
