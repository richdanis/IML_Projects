```python
import pandas as pd
import numpy as np
```

```python
train_str = 'Data/data_Sven_long.csv'
#labels_str = 'train_labels.csv'
df_train = pd.read_csv(train_str)
#df_labels_unsorted = pd.read_csv(labels_str)
```

```python
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', None)
```

```python
df_train = df_train.drop(columns='Time')
df_train
```

```python
new_shape = (df_train.shape[0]//12,df_train.shape[1])
new_shape
```

```python
np_train = np.empty((new_shape))
df_train = df_train.copy()

for i in range(len(df_train.pid.unique())):
    mask = df_train.loc[:,'pid'] == df_train.pid.unique()[i]
    np_train[i,:] = df_train[mask].mean()
    
df_train_short = pd.DataFrame(np_train, columns=df_train.columns)
df_train_short
```

```python
df_train_short = df_train_short.astype({'pid':'int32'})
df_train_short = df_train_short.astype({'Age':'int32'})
df_train_short.to_csv('Data/data_Sven_short.csv', index=False, float_format='%.3f')
```
